### 데이터셋 전처리
- 필터 : 트래블 미션 체크가 있는 경우(not null)
- 트래블 미션 체크를 분리해서 첫번째 유형을 int화

```python
df_filter = df[~df['TRAVEL_MISSION_CHECK'].isnull()].copy()
df_filter.loc[:, 'TRAVEL_MISSION_INT'] = df_filter['TRAVEL_MISSION_CHECK'].str.split(';').str[0].astype(int)
df_filter
```

### 데이터셋 전처리2
- 필요한 열만 추출
- 성별, 나이, 여행 스타일, 여행 목적, 동반자 수, 여행 미션, 방문 지역 이름, 만족도
- 비어있는 행 삭제
- 데이터 개수 3만개 -> 약 만오천개

```python
df_filter = df_filter[[
    'GENDER',
    'AGE_GRP',
    'TRAVEL_STYL_1', 'TRAVEL_STYL_2', 'TRAVEL_STYL_3', 'TRAVEL_STYL_4', 'TRAVEL_STYL_5', 'TRAVEL_STYL_6', 'TRAVEL_STYL_7', 'TRAVEL_STYL_8',
    'TRAVEL_MOTIVE_1',
    'TRAVEL_COMPANIONS_NUM',
    'TRAVEL_MISSION_INT',
    'VISIT_AREA_NM',
    'DGSTFN',
]]

df_filter = df_filter.dropna()
df_filter
```

### 데이터셋 전처리3
- 범주형 데이터 정의(성별, 여행 스타일, 여행 목적, 트래블 미션, 방문지 등등)
- 범주형 데이터는 string or integer 형태이어야 함

```python
categorical_features_names = [
    'GENDER',
    # 'AGE_GRP',
    'TRAVEL_STYL_1', 'TRAVEL_STYL_2', 'TRAVEL_STYL_3', 'TRAVEL_STYL_4', 'TRAVEL_STYL_5', 'TRAVEL_STYL_6', 'TRAVEL_STYL_7', 'TRAVEL_STYL_8',
    'TRAVEL_MOTIVE_1',
    # 'TRAVEL_COMPANIONS_NUM',
    'TRAVEL_MISSION_INT',
    'VISIT_AREA_NM',
    # 'DGSTFN',
]

df_filter[categorical_features_names[1:-1]] = df_filter[categorical_features_names[1:-1]].astype(int)

df_filter
```

### 데이터셋 전처리4
- 학습/테스트 데이터 나누기

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(df_filter, test_size=0.2, random_state=42)

print(train_data.shape)
print(test_data.shape)
```

### CatBoost
- GPU를 사용해서 연산 속도를 빠르게 할 수 있는 게 장점
- 여행객과 여행지의 정보로 이 사람이 이 여행지에 갔을 때 느끼는 만족도 예측

```python
!pip install -q catboost
```

### 데이터 형태 정의
- 만족도를 예측할거기 때문에 feature에 만족도를 제외한 데이터를, label에는 만족도만 입력해서 정의  
- 범주형 데이터 명시

```python
from catboost import CatBoostRegressor, Pool

train_pool = Pool(train_data.drop(['DGSTFN'], axis=1),
                  label=train_data['DGSTFN'],
                  cat_features=categorical_features_names)

test_pool = Pool(test_data.drop(['DGSTFN'], axis=1),
                 label=test_data['DGSTFN'],
                 cat_features=categorical_features_names)
```

## 모델 정의 및 학습
- 캣부스트 회귀모델 사용
- loss function = root min squared error
- evaluation metric = min absolute errora

```python
model = CatBoostRegressor(
    loss_function='RMSE',
    eval_metric='MAE',
    task_type='GPU',
    depth=6,
    learning_rate=0.01,
    n_estimators=2000)

model.fit(
    train_pool,
    eval_set=test_pool,
    verbose=500,
    plot=True)
```

### 서비스 구현 시 참고
- 랜덤한 데이터를 뽑아서 사용자 입력으로 넣는다.
- 여행자의 데이터를 input으로 입력하고 예상 만족도를 구해서 순위 10개를 뽑아준다.

```python
# 여행지 개수 = 8700개
area_names = df_filter[['VISIT_AREA_NM']].drop_duplicates()

area_names
```

```python
traveler = {
    'GENDER': '남',
    'AGE_GRP': 20.0,
    'TRAVEL_STYL_1': 1,
    'TRAVEL_STYL_2': 1,
    'TRAVEL_STYL_3': 2,
    'TRAVEL_STYL_4': 3,
    'TRAVEL_STYL_5': 2,
    'TRAVEL_STYL_6': 2,
    'TRAVEL_STYL_7': 2,
    'TRAVEL_STYL_8': 3,
    'TRAVEL_MOTIVE_1': 8,
    'TRAVEL_COMPANIONS_NUM': 0.0,
    'TRAVEL_MISSION_INT': 3,
}

results = pd.DataFrame([], columns=['AREA', 'SCORE'])

for area in area_names['VISIT_AREA_NM']:
    input = list(traveler.values())
    input.append(area)

    score = model.predict(input)

    results = pd.concat([results, pd.DataFrame([[area, score]], columns=['AREA', 'SCORE'])])

results.sort_values('SCORE', ascending=False)[:10]
```
