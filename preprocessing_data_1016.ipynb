```python
df_place = pd.read_csv('tn_visit_area_info_방문지정보_A.csv', encoding='cp949')
df_place
```

```python
# 방문지 유형 코드 재정의
mapping = {
    1: '자연',    # 자연
    7: '자연',    # 자연
    2: '역사/유적지',    # 역사/유적지
    3: '문화시설',    # 문화시설
    4: '쇼핑',    # 쇼핑
    10: '쇼핑',   # 쇼핑
    5: '레저/스포츠',    # 레저/스포츠
    6: '테마시설',    # 테마시설
    13: '테마시설',   # 테마시설
    11: '음식점/카페'    # 음식점/카페
}

# 지정된 VISIT_AREA_TYPE_CD 코드만 포함하도록 필터링
df_place = df_place[df_place['VISIT_AREA_TYPE_CD'].isin(mapping.keys())]

# 매핑을 적용하여 VISIT_AREA_TYPE_CD를 새 코드로 변경
df_place['VISIT_AREA_TYPE_CD'] = df_place['VISIT_AREA_TYPE_CD'].map(mapping)

df_place
```

```python
# 인덱스 재정렬
df_place = df_place.reset_index(drop = True)

df_place
```

```python
# 방문지 지번주소 & 도로명주소 결측치 제거
df_place.dropna(subset=['ROAD_NM_ADDR', 'LOTNO_ADDR'], how='all', inplace=True)

# 인덱스 재설정
df_place = df_place.reset_index(drop = True)

# 시도/군구 변수 생성
sido = []
gungu = []

for i in range(len(df_place)):
    # LOTNO_ADDR(지번주소) 우선 사용, 결측치면 ROAD_NM_ADDR(도로명주소) 사용
    if pd.notna(df_place['LOTNO_ADDR'][i]):  # 지번주소가 존재하면
        address = df_place['LOTNO_ADDR'][i]
    elif pd.notna(df_place['ROAD_NM_ADDR'][i]):  # 지번주소가 없으면 도로명주소 사용
        address = df_place['ROAD_NM_ADDR'][i]
    else:
        address = None

    if address:  # 주소가 있으면 시도와 군구를 분리
        address_parts = address.split(' ')
        sido.append(address_parts[0])
        gungu.append(address_parts[1] if len(address_parts) > 1 else None)
    else:
        sido.append(None)
        gungu.append(None)

df_place['SIDO'] = sido
df_place['GUNGU'] = gungu

# 시도 통합
df_place['SIDO'] = df_place['SIDO'].replace({
                                              '서울특별시': '서울',
                                              '경기도': '경기',
                                              '인천광역시': '인천',
                                              '강원도': '강원',
                                              '경상남도': '경남',
                                              '전라북도': '전북',
                                              '충청남도': '충남',
                                              '부산광역시': '부산',
                                              '경상북도': '경북',
                                              '대전광역시': '대전',
                                              '제주특별자치도': '제주',
                                              '세종특별자치시': '세종'
                                          })

# 결과 확인
df_place['SIDO'].value_counts()
```

```python
# 도로명 주소, 지번 주소 결측치 채우기

import requests

KAKAO_REST_API_KEY = '949138ef4d7ffbb96c15e3ac9476eb11'

# 도로명 주소 -> 지번 주소
def convert_road_to_lotno(addr):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(addr)
  headers = {"Authorization": "KakaoAK " + KAKAO_REST_API_KEY}
  r = requests.get(url, headers = headers)

  if r.status_code == 200:
    try:
      road_address = r.json().get('documents', [])[0].get('address', None)
      if road_address is not None:
        result = road_address.get('address_name', None)
        if result is None:
          print('지번 주소 없음')
        else:
          print('도로명 주소 지번 주소로 바꾸기 성공 ->'+ result)
      else:
        print('지번 주소가 존재하지 않음')
        return None

    except (IndexError, AttributeError, KeyError):
      print('지번 주소 없음')
      return None

  else:
    return None

  return result

# 지번 주소 -> 도로명 주소
def convert_lotno_to_road(addr):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(addr)
  headers = {"Authorization": "KakaoAK " + KAKAO_REST_API_KEY}
  r = requests.get(url, headers = headers)

  if r.status_code == 200:
    try:
      road_address = r.json().get('documents', [])[0].get('road_address', None)
      if road_address is not None:
        result = road_address.get('address_name', None)
        if result is None:
          print('도로명 주소 없음')
        else:
          print('지번 주소 도로명 주소로 바꾸기 성공 ->'+ result)
      else:
        print('도로명 주소가 존재하지 않음')
        return None

    except (IndexError, AttributeError, KeyError):
      print('도로명 주소 없음')
      return None

  else:
    return None

  return result
```

```python
# 결측치 채우기 함수
def fill_missing_addresses(row):
    if pd.isna(row['ROAD_NM_ADDR']) and pd.notna(row['LOTNO_ADDR']):
        # 도로명 주소가 없고, 지번 주소가 있을 때 도로명 주소 채우기
        print(row['TRAVEL_ID'],row['LOTNO_ADDR'], '도로명 주소 채우기')
        row['ROAD_NM_ADDR'] = convert_lotno_to_road(row['LOTNO_ADDR'])
    elif pd.isna(row['LOTNO_ADDR']) and pd.notna(row['ROAD_NM_ADDR']):
        # 지번 주소가 없고, 도로명 주소가 있을 때 지번 주소 채우기
        print(row['TRAVEL_ID'], row['ROAD_NM_ADDR'], '지번 주소 채우기')
        row['LOTNO_ADDR'] = convert_road_to_lotno(row['ROAD_NM_ADDR'])
    return row

# 데이터프레임에 적용
df_place_seoul = df_place_seoul.apply(fill_missing_addresses, axis=1)

df_place_seoul
```

```python
KAKAO_REST_API_KEY = '949138ef4d7ffbb96c15e3ac9476eb11'

# 주소를 경도와 위도로 변환하는 함수
def get_coordinates_from_address(addr):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(addr)
    headers = {"Authorization": "KakaoAK " + KAKAO_REST_API_KEY}
    r = requests.get(url, headers=headers)

    if r.status_code == 200:
        try:
            # 결과에서 주소 정보 가져오기
            address_info = r.json().get('documents', [])[0].get('address', None)
            if address_info is not None:
                latitude = address_info.get('y', None)  # 위도
                longitude = address_info.get('x', None)  # 경도
                if latitude is None or longitude is None:
                    print('경도 또는 위도 정보 없음')
                else:
                    print(f'주소 -> 경도: {longitude}, 위도: {latitude}')
                    return latitude, longitude
            else:
                print('주소 정보가 존재하지 않음')
                return None, None

        except (IndexError, AttributeError, KeyError):
            print('주소 정보 없음')
            return None, None

    else:
        print('API 요청 실패, 상태 코드:', r.status_code)
        return None, None
```

```python
# 결측치가 있는 경도와 위도를 채우기 위한 함수
def fill_missing_coordinates(row):
    if pd.isnull(row['LONGITUDE']) or pd.isnull(row['LATITUDE']):
        latitude, longitude = get_coordinates_from_address(row['LOTNO_ADDR'])
        if latitude is not None and longitude is not None:
            row['LATITUDE'] = latitude
            row['LONGITUDE'] = longitude
    return row

# apply를 사용해 결측치 채우기
df_place_seoul = df_place_seoul.apply(fill_missing_coordinates, axis=1)

# 결과 출력
df_place_seoul
```
